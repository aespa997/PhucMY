```python
try:
    # Nhập tên của tệp
    print('Enter a class file to grade (i.e. class1 for class1.txt) :', end=' ' )
    nhapten= input()
    # Tên file input và output thực tế 
    filename= nhapten+ '.txt'
    fileketqua = nhapten+ '_grades.txt'
    # Mở file input để đọc
    with open (filename, 'r') as fn:
        print('Successfully opened {}'.format(filename))
        # Tạo mới hoặc reset file output
        with open (fileketqua, 'w') as fkq:
            fkq.write('# this is what {} should look like\n'.format(fileketqua))
        # Gán giá trị các biến
        sodonghople=0
        sodongkhonghople=0
        kituso='0123456789'
        answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
        listdapan= answer_key.split(',')
        dsdiem=[] 
        #Bắt đầu kiểm tra từng dòng của file input
        print('**** ANALYZING ****')
        for line in fn:
            # Gán điểm thi bằng 0
            diem=0
            # Kiểm tra một dòng có đúng số phần tử
            if line.count(',') ==25:
                # Tạo list gồm ID và các đáp án
                line=line.strip('\n')
                danhsach= line.split(',')
                #Kiểm tra ID có hợp lệ
                if len(danhsach[0])==9:
                    for i in range(1,9):
                        if danhsach[0][i] not in kituso:
                            sodongkhonghople+=1
                            print('Invalid line of data: N# is invalid: \n{}'.format(line) )
                            break
                    # Tính điểm khi dòng hợp lệ        
                    else:
                        sodonghople+=1
                        for i in range(25):
                            if danhsach[i+1]=='':
                                diem+=0
                            elif danhsach[i+1]==listdapan[i]:
                                diem+=4
                            else:
                                diem-=1
                        # Cho điểm vào list và viết kết quả vào fileout
                        dsdiem+=[diem]
                        with open (fileketqua, 'a') as fkq:
                            fkq.write('{},{}\n'.format(danhsach[0],diem))
           # Các trường hợp không hợp lệ                                       
                else:
                    sodongkhonghople+=1
                    print('Invalid line of data: N# is invalid: \n{}'.format(line) )
            else:
                sodongkhonghople+=1
                print('Invalid line of data: does not contain exactly 26 values: \n{}'.format(line) )
        # In ra không có lỗi nếu tất cả các dòng đều hợp lệ        
        if sodongkhonghople==0:
            print('No errors found!')
        # In ra số dòng hợp lệ và không hợp lệ
        print('**** REPORT **** \nTotal valid lines of data: {} \nTotal invalid lines of data: {}'.format(sodonghople,sodongkhonghople))     
        # Sắp xếp danh sách điểm theo thứ tự tăng dần
        dsdiemtangdan= sorted(dsdiem)
        # In ra điểm trung bình của cả lớp
        tongdiem=0
        for i in range(sodonghople):
            tongdiem+=dsdiem[i]
        diemtb=tongdiem/sodonghople
        print('Mean (average) score:{}'.format(diemtb))
        # In ra điểm cao nhất
        print('Highest score: {}'.format(dsdiemtangdan[sodonghople-1]))
        # In ra điểm thấp nhất
        print('Lowest score: {}'.format(dsdiemtangdan[0]))
        # In ra điểm trung vị
        print('Range of scores: {}'.format(dsdiemtangdan[sodonghople-1]-dsdiemtangdan[0]))
        if sodonghople%2==1:
            k=(sodonghople-1)/2
            k=int(k)
            print('Median score: {}'.format(dsdiemtangdan[k]))
        else:
            k=sodonghople/2
            k=int(k)
            print('Median score: {}'.format((dsdiemtangdan[k]+dsdiemtangdan[k-1])/2))
# Nếu tên file nhập vào không tồn tại thì báo không tìm thấy file
except FileNotFoundError:
    print('File cannot be found.')
```
