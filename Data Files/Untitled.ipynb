```python

try:
    print('Enter a class file to grade (i.e. class1 for class1.txt) :', end=' ' )
    nhapten= input()
    filename= nhapten+ '.txt'
    with open (filename, 'r') as fn:
        print('Successfully opened {}'.format(filename))
        sodonghople=0
        sodongkhonghople=0
        kituso='0123456789'
        answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
        listdapan= answer_key.split(',')
        danhsachdiem=[]
        print('**** ANALYZING ****')
        for line in fn:
            diem=0
            if line.count(',') ==25:
                danhsach= line.split(',')
                if len(danhsach[0])==9:
                    for i in range(1,9):
                        if danhsach[0][i] not in kituso:
                            sodongkhonghople+=1
                            print('Invalid line of data: N# is invalid: \n{}'.format(line) )
                            break
                    else:
                        sodonghople+=1
                        for j in range(25):
                            if danhsach[j+1]=='':
                                diem=diem
                            if danhsach[j+1]==listdapan[j]:
                                diem+=4
                            else:
                                diem-=1
                        danhsachdiem+=[diem]
                        
                                                  
                else:
                    sodongkhonghople+=1
                    print('Invalid line of data: N# is invalid: \n{}'.format(line) )
            else:
                sodongkhonghople+=1
                print('Invalid line of data: does not contain exactly 26 values: \n{}'.format(line) )
        if sodongkhonghople==0:
            print('No errors found!')
        
        print('**** REPORT **** \nTotal valid lines of data: {} \nTotal invalid lines of data: {}'.format(sodonghople,sodongkhonghople))    
        print(danhsachdiem) 
        danhsachdiemtangdan= sorted(danhsachdiem)
except FileNotFoundError:
    print('File cannot be found.')
```
